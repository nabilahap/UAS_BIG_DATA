```python
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql.functions import year, month, dayofmonth, concat_ws, avg

# Inisialisasi Spark session
spark = SparkSession.builder.appName("TrenHargaGula").getOrCreate()

# Baca data dari CSV ke Spark DataFrame
data = spark.read.csv("gula.csv", header=True, inferSchema=True)

# Ubah tipe data kolom 'Tanggal' menjadi tipe data tanggal
data = data.withColumn("Tanggal", concat_ws("-", year("Tanggal"), month("Tanggal"), dayofmonth("Tanggal")).cast("date"))

# Ambil data ke dalam Pandas DataFrame
data_pandas = data.toPandas()

# Tambahkan plot untuk setiap pasar dan tampilkan satu persatu
markets = data_pandas.columns[1:]  # Ambil nama pasar dari kolom DataFrame kecuali 'Tanggal'

for market in markets:
    # Buat plot untuk pasar saat ini
    plt.figure(figsize=(12, 6))
    
    # Plot garis tren
    sns.regplot(x='Tanggal', y=market, data=data_pandas, scatter_kws={'s': 20}, line_kws={'color': 'red'})
    
    # Atur label dan judul plot
    plt.xlabel('Tanggal')
    plt.ylabel(f'Harga Gula di {market}')
    plt.title(f'Nilai Rata-rata di setiap Titik Data di Pasar {market}')
    
    # Tampilkan nilai rata-rata di setiap titik data
    avg_value = data_pandas[market].mean()
    plt.axhline(y=avg_value, color='green', linestyle='--', label=f'Rata-rata: {avg_value:.2f}')
    
    # Tampilkan plot
    plt.xticks(rotation=45)  # Rotasi label tanggal agar lebih mudah dibaca
    plt.legend()
    plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import month, col
import matplotlib.pyplot as plt
import seaborn as sns

# Inisialisasi Spark session
spark = SparkSession.builder.appName("TrenHargaGula").getOrCreate()

# Baca data dari CSV ke Spark DataFrame
data = spark.read.csv("gula.csv", header=True, inferSchema=True)

# Ubah tipe data kolom 'Tanggal' menjadi tipe data tanggal
data = data.withColumn("Tanggal", data["Tanggal"].cast("date"))

# Tambahkan kolom bulan
data = data.withColumn('Bulan', month('Tanggal'))

# Kelompokkan data berdasarkan bulan dan hitung rata-rata
monthly_data = data.groupBy('Bulan') \
                   .agg({'DKI Jakarta':'avg', 
                         'Pasar Glodok':'avg', 
                         'Pasar Senen Blok III - VI':'avg', 
                         'Pasar Minggu':'avg', 
                         'Pasar Kramat Jati':'avg', 
                         'Pasar Sunter Podomoro':'avg'})

# Konversi DataFrame Spark menjadi Pandas DataFrame
pandas_monthly_data = monthly_data.toPandas()

# Plot heatmap menggunakan Matplotlib dan Seaborn
plt.figure(figsize=(12, 6))
sns.heatmap(pandas_monthly_data.set_index('Bulan'), cmap='YlGnBu', annot=True, fmt=".2f")
plt.xlabel('Pasar')
plt.ylabel('Bulan')
plt.title('Rata-rata Harga Gula per Bulan di Berbagai Pasar')
plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

# Inisialisasi Spark session
spark = SparkSession.builder.appName("TrenHargaGula").getOrCreate()

# Baca data dari CSV ke Spark DataFrame
data = spark.read.csv("gula.csv", header=True, inferSchema=True)

# Hapus baris dengan nilai NULL
df = data.dropna()

# Ubah kolom Tanggal ke dalam bentuk timestamp
df = df.withColumn("Tanggal", df["Tanggal"].cast("timestamp"))

# Gabungkan fitur-fitur yang ingin Anda gunakan dalam satu kolom vektor
feature_cols = ['Pasar Glodok', 'Pasar Senen Blok III - VI', 'Pasar Minggu', 'Pasar Kramat Jati', 'Pasar Sunter Podomoro']
assembler = VectorAssembler(inputCols=feature_cols, outputCol='features')
df = assembler.transform(df)

# Pisahkan data menjadi set pelatihan dan set pengujian
train_data, test_data = df.randomSplit([0.8, 0.2], seed=1234)

# Inisialisasi model regresi linear
lr = LinearRegression(featuresCol='features', labelCol='DKI Jakarta')

# Latih model
lr_model = lr.fit(train_data)

# Prediksi harga pada set pengujian
test_results = lr_model.transform(test_data)

# Evaluasi menggunakan RegressionEvaluator
evaluator = RegressionEvaluator(labelCol='DKI Jakarta', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(test_results)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

# Evaluasi menggunakan RegressionEvaluator
evaluator = RegressionEvaluator(labelCol='DKI Jakarta', predictionCol='prediction')

# Menghitung MSE
mse = evaluator.evaluate(test_results, {evaluator.metricName: "mse"})
print("Mean Squared Error (MSE) on test data = %g" % mse)

# Menghitung MAE
mae = evaluator.evaluate(test_results, {evaluator.metricName: "mae"})
print("Mean Absolute Error (MAE) on test data = %g" % mae)

# Menghitung R-squared (R2)
r2 = evaluator.evaluate(test_results, {evaluator.metricName: "r2"})
print("R-squared (R2) on test data = %g" % r2)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ambil hasil prediksi dan konversi ke Pandas DataFrame untuk memudahkan visualisasi
lr_predictions_pd = lr_model.transform(test_data).select("Tanggal", "DKI Jakarta", "prediction").toPandas()

# Buat scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(lr_predictions_pd["DKI Jakarta"], lr_predictions_pd["prediction"], alpha=0.5, color='blue')
plt.title('Linear Regression: Actual vs. Predicted Prices')
plt.xlabel('Harga Aktual (DKI Jakarta)')
plt.ylabel('Harga Prediksi')
plt.show()
```
